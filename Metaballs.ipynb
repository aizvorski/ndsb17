```python
import numpy as np
from pylab import *
%matplotlib inline
```

```python
import scipy.ndimage.filters
```

```python
vsize = np.asarray((32,32,32))
grid = np.indices(vsize).astype(np.float32)
field = np.zeros(vsize)

for n in range(5):
    center = np.random.uniform(8,24,3)
    g = grid - center[:,None,None,None]
    field += np.exp(- 0.2 * np.sqrt(np.sum(np.square(g), axis=0)) )
    print(center)
```

```python
result = (field > 1).astype(np.float32)
result = scipy.ndimage.filters.gaussian_filter(result, sigma=0.7)
```

```python
for i in range(10,22):
    imshow(result[i])
    show()
```
